### Imports

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier
import warnings

warnings.filterwarnings('ignore')
```

### Importando dados

```python
df = pd.read_excel('../datasets/Chamados_Completo_BOW.xlsx')
display(df.info())
df.head()
```

- Labels

```python
y = df['label']
y.head()
```

- Features

```python
X = df.drop(columns=['label'])
X.head()
```

### Validação cruzada

Os valores resultantes do cross_val_score serão aplicados no [Teste de Friedman](https://www.socscistatistics.com/tests/friedman/default.aspx) para análise dos resultados dos modelos

- Modelos

```python
# Definindo modelos
names = ["k-Nearest Neighbors", "Árvores de decisão", "Naive Bayes", "MultiLayer Perceptron (MLP)"]

classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    MultinomialNB(),
    MLPClassifier()
]

models = zip(names, classifiers)
models
```

- Cross validation

```python
for name, model in models:
    print(name, '\n')
    scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')
    scores_rounded = [round(s, 4) for s in scores]
    print("Accuracy scores: ", scores_rounded, '\n')
    print('========================================================')
```
